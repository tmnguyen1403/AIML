```python
import sys
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generate 2D meshgrid
nx, ny = (100, 100)
x = np.linspace(0,10,nx) #
#print(x)
y = np.linspace(0,10,ny)
xv, yv = np.meshgrid(x,y)
```

```python
#function to plot
def f(x,y):
    return x * (y**2)

# calculate Z valu for each x,y point
z = f(xv,yv)
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z)
plt.title('2D Color Plot of f(x,y) = xy^2')
plt.colorbar()
plt.show()
```

```python
# generate 2D meshgrid for the gradient
nx, ny = (10,10)
x = np.linspace(0,10,nx)
y = np.linspace(0,10,ny)
xg, yg = np.meshgrid(x,y)
```

```python
# Calculate the gradient of f(x,y)
# np returns row first, then colums
# since rows are y and columns are x
Gy, Gx = np.gradient(f(xg,yg))
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z)
plt.title('Gradient of f(x,y) = xy^2')
plt.colorbar()
plt.quiver(xg,yg, Gx, Gy, scale = 1000, color = 'w')
plt.show()
```

```python
# Calculate the gradient of f(x,y)
# np returns row first, then colums
# since rows are y and columns are x
def ddx(x,y):
    return y**2
def ddy(x,y):
    return 2*x*y
Gxm, Gym = ddx(xg,yg), ddy(xg,yg)
```

```python
# make a color plot to display the data
plt.figure(figsize=(14,12))
plt.pcolor(xv,yv,z)
plt.title('Plot of [y^2,2xy]')
plt.colorbar()
plt.quiver(xg,yg, Gxm, Gym, scale = 1000, color = 'w')
plt.show()
```
